```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

# Create main window
root = tk.Tk()
root.title("Identification Form")
root.geometry("500x200")

# Create lastname label and entry
lastname_label = tk.Label(root, text="Lastname:")
lastname_label.pack()
lastname_entry = tk.Entry(root)
lastname_entry.pack()

# Create department label and entry
department_label = tk.Label(root, text="Department:")
department_label.pack()
department_entry = tk.Entry(root)
department_entry.pack()

# function for welcomemessage for acct dept
def welcomeMessageAcct(lastname):
    # Create a Tkinter window
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x500")

    label1_1 = tk.Label(window, text="ACCOUNTING DEPARTMENT EMPLOYEES:", anchor= 'w')
    label1_1.pack(fill= 'x')
    
    label1_1 = tk.Label(window, anchor= 'w', text="ADEWUMI\n AGBAKURU-NWOGU\n AKINDE\n EBI\n ICHA\n IKPATI\n OBASOGIE\n OBI\n OBIALOR\n OLOWOKERE\n OLUBUADE\n OSEMEKE")
    label1_1.pack(fill= 'x')

# function for welcomemessage for logistics dept
def welcomeMessageLog(lastname):
    # Create a Tkinter window
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x500")

    label1_1 = tk.Label(window, text="LOGISTICS DEPARTMENT EMPLOYEES:", anchor= 'w')
    label1_1.pack(fill= 'x')
    
    label1_1 = tk.Label(window, anchor= 'w', text="ADENIRAN\n AGBAKWURU\n ARNIKA\n CHUKWUMA\n EZEONYE\n GIWA\n NWOKOLO\n NWOTOKUBO\n OKORO\n OSSAI\n PETER")
    label1_1.pack(fill= 'x')

# function for welcomemessage for admin dept
def welcomeMessageAdmin(lastname):
    # Create a Tkinter window
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x500")

    label1_1 = tk.Label(window, text="ADMINISTRATION DEPARTMENT EMPLOYEES:", anchor= 'w')
    label1_1.pack(fill= 'x')
    
    label1_1 = tk.Label(window, anchor= 'w', text="EGBORO\n ELERI\n EZE\n OKOCHA-OJEAH\n OKOLO\n UMEH")
    label1_1.pack(fill= 'x')

# function for welcomemessage for delivery dept
def welcomeMessageDel(lastname):
    # Create a Tkinter window
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x500")

    label1_1 = tk.Label(window, text="DELIVERY DEPARTMENT EMPLOYEES:", anchor= 'w')
    label1_1.pack(fill= 'x')
    
    label1_1 = tk.Label(window, anchor= 'w', text="ADOH-AJAGBE\n ATELLY\n AZOGU\n BETURE\n EJEDIMU\n ILOENYOSI\n IYAWE\n OGBONNA\n OIGBOCHIE\n QUADRI\n UDE-IBE")
    label1_1.pack(fill= 'x')

# submit button callback function
def submit():
    lastname = lastname_entry.get()
    department = department_entry.get()

    if department == "Accounting":
        if lastname in ["Adewumi", "Agbakuru-Nwogu", "Akinde", "Ebi", "Icha", "Ikpati", "Obasogie", "Obi", "Obialor", "Olowokere", "Olubuade", "Osemeke"]:
            messagebox.showinfo("Success", f"Welcome, {lastname}!")
            welcomeMessageAcct(lastname)
        else:
            messagebox.showerror("Login", "Unable to identify user")

    elif department == "Logistics":
        if lastname in ["Adeniran", "Agbakwuru", "Arnika", "Chukwuma", "Ezeonye", "Giwa", "Nwokolo", "Nwotokubo", "Okoro", "Ossai", "Peter"]:
            messagebox.showinfo("Success", f"Welcome, {lastname}!")
            welcomeMessageLog(lastname)
        else:
            messagebox.showerror("Login", "Unable to identify user")

    elif department == "Administration":
        if lastname in ["Egboro", "Eleri", "Eze", "Okocha-Ojeah", "Okolo", "Umeh"]:
            messagebox.showinfo("Success", f"Welcome, {lastname}!")
            welcomeMessageAdmin(lastname)
        else:
            messagebox.showerror("Login", "Unable to identify user")

    elif department == "Delivery":
        if lastname in ["Adoh-Ajagbe", "Atelly", "Azogu", "Beture", "Ejedimu", "Iloenyosi", "Iyawe", "Ogbonna", "Oigbochie", "Quadri", "Ude-Ibe"]:
            messagebox.showinfo("Success", f"Welcome, {lastname}!")
            welcomeMessageDel(lastname)
        else:
            messagebox.showerror("Login", "Unable to identify user")

 # Create submit button
submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

# Run the main event loop
root.mainloop()
```
